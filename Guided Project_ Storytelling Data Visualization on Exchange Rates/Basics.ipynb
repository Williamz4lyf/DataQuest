### Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar.loc[euro_to_dollar['US_dollar'] == '-']
```

```python
euro_to_dollar.drop(euro_to_dollar[euro_to_dollar['US_dollar'] == '-'].index, inplace=True)
```

```python

euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

### Rolling Mean

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

When we review the line plot above, we see tiny variations representing daily changes in the euro to dollar exchange rate. These change are generally superficial and significant changes are only seen over time.

In order to hide these timy variations and highlight only long term trends, we use a rolling mean (also known as a moving average). A rolling mean / moving average is the mean between the value of one day and the value of the previous day. We can calculate the mean for each day using the pandas.Series.rolling().mean() method.

 The number of days used in the calculation of the rolling mean is called the rolling window (or moving window). The rolling window can take various values.

We will calculate the rolling mean of the US_dollar column using a moving window of 30 days.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar.head(100)
```

### Plot Rolling Mean for 7, 30, 50, 100 and 365-day Windows.

```python
plt.figure(figsize=(10,6))
plot_num = list(range(2,7))
roll_win = [7, 30, 50, 100, 365]

plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.title('Original Values', weight='bold')

for num, window in zip(plot_num, roll_win):
    plt.subplot(3, 2, num)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(window).mean())
    plt.title(f'Rolling Window: {window}', weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

### Deciding on the Graph to Build

There are a number of story ideas for this data including:
1. We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data and the 2016-2019 data as a baseline. We can use a line plot.
2. We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 and 2009 for comparison. We can use a line plot.
3. We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

```python
# EUR-USD Financial Crisis df
fin_range = euro_to_dollar.loc[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]
fin_crisis = euro_to_dollar.loc[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
# EUR-USD Financial Crisis Plot
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12,5))

# Plot the line chart
ax.plot(fin_range['Time'], fin_range['rolling_mean'],
        color='#15B01A', alpha=0.3)

# Remove the borders
for location in ['left', 'right', 'bottom', 'top']:
    ax.spines[location].set_visible(False)

# Add grid
ax.grid(visible=True)

# Remove certain ygridlines
ygridlines = ax.get_ygridlines()
for index, gridline in enumerate(ygridlines):
    if index % 2 == 0:
        gridline.set_visible(False)

# Remove borders and set tick labels
ax.tick_params(left=False, bottom=False)
ax.set_yticklabels(['', 1.2, '', 1.3, '', 1.4, '', 1.5],
                   size=12, color='grey')
ax.set_xticklabels([])
x_coords = {'2006':13120, '2007':13485, '2008':13850,
            '2009':14216, '2010':14581}
y_coord = 1.135
for x_label, x_coord in x_coords.items():
    ax.text(x_coord, y_coord, x_label, size=12, color='grey')

# Plot Red Line Chart
ax.plot(fin_crisis['Time'], fin_crisis['rolling_mean'],
         color='#FF0000', linewidth=3.5, alpha=0.7)

# Title, Subtitle, Footer
ax.text(13000, 1.06, 'Â©DATAQUEST' + ' '*115 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=12)
ax.text(13000, 1.65,
        'Euro-USD rate peaked at 1.59 during 2007-2008\'s financial crisis',
        size=19, weight='bold')
ax.text(13000, 1.615,
        'Euro-USD exchange rates between 2006 and 2010', size=18)

# Plot Shaded Bar
ax.axvline(x=14050, ymin=0.25, c='grey', linewidth=95, alpha=0.3)

plt.show()
```
